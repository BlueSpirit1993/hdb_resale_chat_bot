```python
town_region_mapping = {
    "TAMPINES": "East",
    "YISHUN": "North",
    "JURONG WEST": "West",
    "BEDOK": "East",
    "WOODLANDS": "North",
    "ANG MO KIO": "Central",
    "HOUGANG": "North-East",
    "BUKIT BATOK": "West",
    "CHOA CHU KANG": "West",
    "BUKIT MERAH": "Central",
    "SENGKANG": "North-East",
    "PASIR RIS": "East",
    "TOA PAYOH": "Central",
    "QUEENSTOWN": "Central",
    "GEYLANG": "Central",
    "CLEMENTI": "West",
    "BUKIT PANJANG": "West",
    "KALLANG/WHAMPOA": "Central",
    "JURONG EAST": "West",
    "SERANGOON": "North-East",
    "PUNGGOL": "North-East",
    "BISHAN": "Central",
    "SEMBAWANG": "North",
    "MARINE PARADE": "East",
    "CENTRAL AREA": "Central",
    "BUKIT TIMAH": "Central",
    "LIM CHU KANG": "West",
}

df["region"] = df["town"].map(town_region_mapping)
```

```python
# low level = 1 to 9
# Mid level = 10 to 19
# High level = 20 & above

storey_mapping = {
    "10 TO 12": "mid_level(10 to 19)",
    "04 TO 06": "low_level(1 to 9)",
    "01 TO 03": "low_level(1 to 9)",
    "13 TO 15": "mid_level(10 to 19)",
    "16 TO 18": "mid_level(10 to 19)",
    "07 TO 09": "low_level(1 to 9)",
    "19 TO 21": "high_level(20 & above)",
    "22 TO 24": "high_level(20 & above)",
    "25 TO 27": "high_level(20 & above)",
    "28 TO 30": "high_level(20 & above)",
    "31 TO 33": "high_level(20 & above)",
    "37 TO 39": "high_level(20 & above)",
    "40 TO 42": "high_level(20 & above)",
    "34 TO 36": "high_level(20 & above)",
    "01 TO 05": "low_level(1 to 9)",
    "06 TO 10": "low_level(1 to 9)",
    "11 TO 15": "mid_level(10 to 19)",
    "16 TO 20": "mid_level(10 to 19)",
    "26 TO 30": "high_level(20 & above)",
    "36 TO 40": "high_level(20 & above)",
    "21 TO 25": "high_level(20 & above)",
    "31 TO 35": "high_level(20 & above)",
    "46 TO 48": "high_level(20 & above)",
    "43 TO 45": "high_level(20 & above)",
    "49 TO 51": "high_level(20 & above)",
}
df["level"] = df["storey_range"].map(storey_mapping)
```

```python
df["floor_area_sqm"] = df["floor_area_sqm"].astype("int32")
```

```python
df.lease_commence_date.unique().min(), df.year_of_sales.unique().min()
```

```python
df.head()
```

```python
# encode year of sales from 1990 and lease commence date from 1966
df["salesYear_fr_1990"] = df.year_of_sales - 1990
df["leaseYear_fr_1966"] = df.lease_commence_date - 1966
```

```python
# encodes a month of sales using sine and cosine transformations."""
df["month_sin"] = np.sin((df.month_of_sales - 1) * (2 * np.pi / 12))
df["month_cos"] = np.cos((df.month_of_sales - 1) * (2 * np.pi / 12))
```

```python
# Instantiate the Ordinal Encoder
ordinal_encoder = OrdinalEncoder(
    categories=[["low_level(1 to 9)", "mid_level(10 to 19)", "high_level(20 & above)"]]
)

# Fit it
ordinal_encoder.fit(df2[["level"]])

# Display the learned categories
display(ordinal_encoder.categories_)
```

```python
# Transform categories into ordered numbers
df2["encoded_level"] = ordinal_encoder.transform(df2[["level"]])
```

## Training model

```python
from sklearn.linear_model import (
    SGDRegressor,
    LinearRegression,
    Ridge,
    ElasticNet,
    Lasso,
)
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.dummy import DummyRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)
```

```python
baseline_model = DummyRegressor(strategy="mean")  # Baseline
baseline_model.fit(X_train, y_train)  # Calculate value for stratgy
baseline_model.score(
    X_test, y_test
)  # R2 Score model based on consistently predicting the strategy
```

```python
# train linear reg model without CV
```

```python
# train linear reg model with CV
model_lingreg_cv = LinearRegression()
cross_val_score(model_lingreg_cv, X_train, y_train, cv=5)
```

```python
# train SGD reg model with CV

model_SGD_cv = SGDRegressor()
cross_val_score(model_SGD_cv, X_train, y_train, cv=5)
```

```python
# train ridge model with CV

model_ridge_cv = Ridge()
cross_val_score(model_ridge_cv, X_train, y_train, cv=5)
```

```python
# train elastic net model with CV

model_elastic_cv = ElasticNet()
cross_val_score(model_elastic_cv, X_train, y_train, cv=5)
```

```python
# train Lasso model with CV

model_lasso_cv = Lasso()
cross_val_score(model_lasso_cv, X_train, y_train, cv=5)
```
